```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
```

```python
# función en la que se va a optimizar
func = lambda th: np.sin(1/2 * th[0]**2 - 1/4 * th[1]**2 + 3) * np.cos(2*th[0]+1-np.e**th[1])

res = 100

_x = np.linspace(-2,2,res) # rango de la grafica en x
_y = np.linspace(-2,2,res) # rango de la grafica en y

_z = np.zeros((res,res))

for ix,x in enumerate(_x):
    for iy, y in enumerate(_y):
        _z[iy,ix] = func([x,y])
        
plt.contourf(_x, _y, _z, 100)
plt.colorbar()

Theta = np.random.rand(2) * 4 - 2 # se inicia en una posición aleatoria. Se da formato de -2-2

_t = np.copy(Theta)

h = 0.001 # variación
lr = 0.001 # razón de aprendizaje

plt.plot(Theta[0], Theta[1], "o", c = "white") 

grad = np.zeros(2)

for _ in range(1000):
    for it, th in enumerate(Theta):
        _t = np.copy(Theta)
        _t[it] = _t[it] + h
        deriv = (func(_t) - func(Theta)) / h
        grad[it] = deriv

    Theta = Theta - lr * grad
    
    if(_ % 100 == 0):
        plt.plot(Theta[0], Theta[1], ".", c = "red")

plt.plot(Theta[0], Theta[1], "o", c = "green")
plt.show()
```
